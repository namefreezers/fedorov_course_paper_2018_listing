```python
import numpy as np
import pandas as pd
import re
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.linear_model import LogisticRegression
# import nltk
# nltk.download()
from nltk.stem import WordNetLemmatizer
```

```python
df_train = pd.read_csv('train.csv', index_col='id')
# df_train = df_train.sample(frac=0.1)
df_train.head()
```

```python
df_test = pd.read_csv('test.csv', index_col='id')
# df_test = df_test.sample(frac=0.1)
df_test.head()
```

```python
re_not_words = re.compile("[^\w']+")
for df in [df_train, df_test]:
    df.comment_text = df.comment_text.apply(lambda x: re_not_words.sub(' ', x))
df_train.head()
```

```python
for x in [df_train, df_test]:
    x.comment_text = x.comment_text.apply(lambda x: x.lower().strip())
df_train.head()
```

```python
wnl = WordNetLemmatizer()
for x in [df_train, df_test]:
    x.comment_text = x.comment_text.apply(lambda x: ' '.join([wnl.lemmatize(token) for token in re.split('\s+', x)]))
df_train.head()
```

```python
vctrzr = TfidfVectorizer(ngram_range=(1, 1), stop_words='english', max_features=80_000)
vctrzr.fit(df_train.comment_text.append(df_test.comment_text))
train_m = vctrzr.transform(df_train.comment_text).astype(np.float32)
test_m = vctrzr.transform(df_test.comment_text).astype(np.float32)
del vctrzr
train_m, test_m
```

```python
vctrzr_chars = TfidfVectorizer(analyzer='char', ngram_range=(3, 5), max_features=40_000)
vctrzr_chars.fit(df_train.comment_text.append(df_test.comment_text))
train_chars_m = vctrzr_chars.transform(df_train.comment_text).astype(np.float32)
test_chars_m = vctrzr_chars.transform(df_test.comment_text).astype(np.float32)
del vctrzr_chars
train_chars_m, test_chars_m
```

```python
import scipy
train_words_chars_m = scipy.sparse.hstack((train_m, train_chars_m), format='csr')
test_words_chars_m = scipy.sparse.hstack((test_m, test_chars_m), format='csr')
```

```python
clf_toxic = LogisticRegression(penalty='l2')
clf_toxic.fit(train_words_chars_m, df_train.toxic)
df_test['toxic'] = clf_toxic.predict_proba(test_words_chars_m)[:,1]

clf_severe_toxic = LogisticRegression(penalty='l2')
clf_severe_toxic.fit(train_words_chars_m, df_train.severe_toxic)
df_test['severe_toxic'] = clf_severe_toxic.predict_proba(test_words_chars_m)[:,1]

clf_obscene = LogisticRegression(penalty='l2')
clf_obscene.fit(train_words_chars_m, df_train.obscene)
df_test['obscene'] = clf_obscene.predict_proba(test_words_chars_m)[:,1]

clf_threat = LogisticRegression(penalty='l2')
clf_threat.fit(train_words_chars_m, df_train.threat)
df_test['threat'] = clf_threat.predict_proba(test_words_chars_m)[:,1]

clf_insult = LogisticRegression(penalty='l2')
clf_insult.fit(train_words_chars_m, df_train.insult)
df_test['insult'] = clf_insult.predict_proba(test_words_chars_m)[:,1]

clf_identity_hate = LogisticRegression(penalty='l2')
clf_identity_hate.fit(train_words_chars_m, df_train.identity_hate)
df_test['identity_hate'] = clf_identity_hate.predict_proba(test_words_chars_m)[:,1]
```
